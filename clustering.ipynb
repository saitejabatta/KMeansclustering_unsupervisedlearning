```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
data=pd.read_csv("penguins1.csv")
x=data.iloc[:,9:11].values
plt.scatter(x[:,0],x[:,1],marker='x')
plt.grid(True)
plt.title("Scatter plot")
plt.legend()
plt.show()
wcss=[]
for i in range(1,11):
    k=KMeans(n_clusters=i,init='k-means++',random_state=0)
    k.fit(x)
    wcss.append(k.inertia_)
plt.plot(range(1,11),wcss)
plt.scatter(range(1,11),wcss,c='k')
plt.title("elbow graph")


k=KMeans(n_clusters=3,init='k-means++',random_state=0)
y=k.fit_predict(x)
y
k.cluster_centers_




    
```

```python
plt.scatter(x[y==0,0],x[y==0,1],c='k',label='cluster1')
plt.scatter(x[y==1,0],x[y==1,1],c='r',label='cluster2')
plt.scatter(x[y==2,0],x[y==2,1],c='b',label='cluster3')
plt.legend()
plt.title("KMeans_clustering")
plt.xlabel("xxxx")
plt.ylabel("yyyy")
plt.show()
```
